```python
# --- GDP Data Visualization for India ---
# Run this in Google Colab

# Step 1: Install necessary packages
!pip install pandas_datareader matplotlib seaborn

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_datareader import wb

# Step 3: Fetch GDP data from World Bank
# GDP indicator code: "NY.GDP.MKTP.CD" (GDP in current US$)
gdp_data = wb.download(indicator='NY.GDP.MKTP.CD', country='IN', start=1960, end=2024)
gdp_data.reset_index(inplace=True)

# Step 4: Rename columns for clarity
gdp_data.rename(columns={'NY.GDP.MKTP.CD': 'GDP (US$)', 'year': 'Year'}, inplace=True)

# Step 5: Plot the GDP trend
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='GDP (US$)', data=gdp_data, marker='o', color='blue')
plt.title('GDP Trend of India (19602024)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('GDP (US$)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 6: Display last 10 years for reference
print("Recent GDP data for India:")
display(gdp_data.tail(10))
```
